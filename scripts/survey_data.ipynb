# Live survey plots

Step 1: import the **pandas** library, which contains subroutines for managing data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Step 2: Collect the data. Save the data to a pandas **dataframe**, which acts a lot like a spreadsheet, storing data in rows and columns. Our survey results can be stored in a very simple, 1-column dataframe.

Functions in the pandas library are called using the `pd` prefix we assigned it when we imported. *Italics*

```python
Enthusiasm = {
    '1': [4,3,5,2],
    '2': [4,9,4,8],
    '3': [1,4,6,3],
    '4': [3,2,1,1],
    '5': [0,0,1,0]
}
dfEnthusiasm = pd.DataFrame(Enthusiasm, index=[2021,2022,2023,2024]).T
```

```python
dfEnthusiasm
```

Step 3: Plot!

The `dEnthusiasm` dataframe is an **object** (or **class** in Pythonese), that is a chunk computer code that contains the data as well as functions (called **methods**) that can be applied to the data.

You can do a lot with dataframes.

```python
dfEnthusiasm.plot();
```

Maybe a bar plot is easier to read?

```python
StudentCount = dfEnthusiasm.sum()
StudentCount
```

```python
(dfEnthusiasm/StudentCount).plot.bar();
plt.ylabel("Fraction of first-years")
plt.xlabel("Enthusiam (1=highest, 5=lowest)")
```

```python
dfEnthusiasm.describe()
```

Different packages can make prettier plots. For example **seaborn** is a very popular plotting library. **Matplotlib** is a lower-level plotting library that forms the foundation of seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ApplicationFrac = dfApplications/StudentCount
ApplicationFrac = ApplicationFrac.reset_index().melt(id_vars=['index'])
ApplicationFrac
```

```python
sns.barplot(data=ApplicationFrac, x='index', y='value', hue='variable')#, x=dfApplications.index, y=[[2021,2022,2023]])
plt.xlabel('Applications');
plt.xticks(rotation=45);
```
